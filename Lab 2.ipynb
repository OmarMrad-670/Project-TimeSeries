# Modlisation

```r
VD_36<-ifelse(time>t1 & time<=t2,1,0)

VD_83<-ifelse(time>t2 & time<=t3,1,0)

VD_120<-ifelse(time>t3 & time<=t4,1,0)

VD_204<-ifelse(time>t4,1,0)
```

```r
time<-c(1:length(Japan.ts))
```

```r
model1<-lm(Japan.ts~time) #Y=Japan.ts la variable explicative et X=time variable descriptive
model2<-lm(Japan.ts~time+I(time^2))

model3<-lm(Japan.ts~time*(VD_36+VD_83+VD_120+VD_204))
model4<-lm(Japan.ts~time*(VD_36+VD_83+VD_120+VD_204)+I(time^2)*(VD_36+VD_83+VD_120+VD_204))
```

```r
summary(model1)
summary(model2)
summary(model3)
summary(model4)
```

```r
plot(Japan.ts)
points(time(Japan.ts),fitted(model1),col="red",type="l",lwd=2)
points(time(Japan.ts),fitted(model2),col="green",type="l",lwd=2)
points(time(Japan.ts),fitted(model3),col="blue",type="l",lwd=2)
points(time(Japan.ts),fitted(model4),col="yellow",type="l",lwd=2)
```

```r
AIC(model1,model2,model3,model4)
```

```r
BIC(model1,model2,model3,model4)
```

## Modlisation de la composante stochastique (rsiduelle)

#### Reprsentation graphique de la composante rsiduelle

```r
Japan.res=model14$residuals
```

```r
plot(time(Japan.res),Japan.res, xlab="Temps", ylab="",main="La composante rsiduelle",col="purple3",lwd=2,type="l")
```

### La courbe est sinusoidale , donc il n'a pas de bruit Blanc 

```r
acf(Japan.res)
```

### D'apres ACF ,il y'a depassement de seuil donc on a des informations donc le bruit il n'est pas blanc 

### Test de stationnarit

```r
library(tseries)
```

```r
adf.test(Japan.res)
```

#### La p_value= 0.01 <0.05  => donc on accepte lhypothse alternative : les donnes sont stationnaire.

### Test de blancheur

```r
Box.test(Japan.res)
```

La valeurs de la p_value= p-value = 2.2e-16 <0.05 que lhypothse alternative est considre : le processus stationnaire ne se comporte pas comme bruit blanc.
