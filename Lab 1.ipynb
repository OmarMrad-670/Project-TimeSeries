# Modélisation

```r
VD_58=ifelse(time>=58,1,0)
```

```r
time<-c(1:length(Bureau.ts))
```

```r
model1<-lm(Bureau.ts~time) #Y=Bureau.ts la variable explicative et X=time variable descriptive
model2<-lm(Bureau.ts~time+I(time^2))
model3<-lm(Bureau.ts~time*VD_58)
model4<-lm(Bureau.ts~time*VD_58+I(time^2)*VD_58)
```

```r
summary(model1)
summary(model2)
summary(model3)
summary(model4)
```

```r
plot(Bureau.ts)
points(time(Bureau.ts),fitted(model1),col="red",type="l",lwd=2)
points(time(Bureau.ts),fitted(model2),col="green",type="l",lwd=2)
points(time(Bureau.ts),fitted(model3),col="blue",type="l",lwd=2)
points(time(Bureau.ts),fitted(model4),col="yellow",type="l",lwd=2)
```

```r
AIC(model1,model2,model3,model4)
```

```r
BIC(model1,model2,model3,model4)
```

# Modélisation de la tendance et de la composante saisonnière : 

## T= 12

```r
library(forecast)
DV<-seasonaldummy(Bureau.ts) # a matrix composed of all dummy variables
```

```r
MC<-matrix(0,length(Bureau.ts),6)#6 variables descriptive
MS<-matrix(0,length(Bureau.ts),6)
for (i in 1:6){MC[,i]<- cos(2*pi*i*time/12) }
for (i in 1:6){MS[,i]<- sin(2*pi*i*time/12) }

#### Saisonalite normale
model11<-lm(Bureau.ts~time+MC+MS)
model12<-lm(Bureau.ts~time+I(time^2)+MC+MS)

### DV
model13<-lm(Bureau.ts~time*VD_58+DV)
model14<-lm(Bureau.ts~time*VD_58+I(time^2)*VD_58+DV)


#### Saisonalite MC,MS
model15<-lm(Bureau.ts~time*VD_58+MC+MS)
model16<-lm(Bureau.ts~time*VD_58+I(time^2)*VD_58+MC+MS)
```

```r
summary(model11)
summary(model12)
summary(model13)

summary(model14)
summary(model15)
summary(model16)
```

```r
AIC(model11,model12,model13,model14,model15,model16)
```

```r
BIC(model11,model12,model13,model14,model15,model16)
```

### Donc on choisit le 15eme modele avec Adjusted R-squared= 0.9727

```r
#Représentation graphique du modèle
plot(Bureau.ts)
points(time(Bureau.ts),fitted(model13),col="blue",type="l",lwd=2)
points(time(Bureau.ts),fitted(model15),col="red",type="l",lwd=2)
```

## Modélisation de la composante stochastique (résiduelle)

#### Représentation graphique de la composante résiduelle

```r
Bureau.res=model15$residuals
```

```r
plot(time(Bureau.res),Bureau.res, xlab="Temps", ylab="",main="La composante résiduelle",col="purple3",lwd=2,type="l")
```

### La courbe est sinusoidale , donc il n'a pas de bruit Blanc 

```r
acf(Bureau.res)
```

### D'apres ACF ,il y'a depassement de seuil donc on a des informations donc le bruit il n'est pas blanc 

### Test de stationnarité

```r
library(tseries)
```

```r
adf.test(Bureau.res)
```

#### La p_value= 0.01 <0.05  => donc on accepte l’hypothèse alternative : les données sont stationnaire.

### Test de blancheur

```r
Box.test(Bureau.res)
```

La valeurs de la p_value= p-value = 0.01073 <0.05 que l’hypothèse alternative est considérée : le processus stationnaire ne se comporte pas comme bruit blanc.

## Modèle ARMA pour la composante résiduelle

```r
par(mfrow=c(1,2))
acf(Bureau.res,120)#### q
pacf(Bureau.res,120)### p
```

#### On va utiliser la fonction auto.arima() du package forecast afin de determiner l’ordre p du modèle AR et l’ordre q de celui de MA, vu que c’est difficile de les savoir à partir de l’acf et la pacf.

```r
library(forecast)
```

```r
ARIMA=auto.arima(Bureau.res,d=1,trace=TRUE)
```

```r
Model.res<-arima(Bureau.res,order = c(2, 1, 0))
```

```r
adf.test(Model.res$residuals)
```

## Puisque p_value = 0.01 < 0.05 Donc Stationnaire

```r
checkresiduals(Model.res)
```

```r
d.Bureau.ts.12<-diff(Bureau.res,12,1)
acf(d.Bureau.ts.12,50)
pacf(d.Bureau.ts.12,50)
```

```r
arima(Bureau.res, order = c(2, 1, 0),seasonal = list(order = c(0,1,2), period = 12))$sigma2
arima(Bureau.res, order = c(2, 1, 0),seasonal = list(order = c(1,1,2), period = 12))$sigma2 
arima(Bureau.res, order = c(2, 1, 0),seasonal = list(order = c(2,1,0), period = 12))$sigma2
arima(Bureau.res, order = c(2, 1, 0),seasonal = list(order = c(2,1,1), period = 12))$sigma2 #the selected model
```

```r
Model.res.s<-arima(Bureau.res, order = c(2, 1, 0),seasonal = list(order = c(2,1,1), period = 12))
adf.test(Model.res.s$residuals)
```

```r
summary(Model.res.s)
checkresiduals(Model.res.s)
```

# Forecasting using only SARIMA model
We apply sarima model on the original time series

```r
Model.SARIMA<-auto.arima(Bureau.ts)
summary(Model.SARIMA)
```

```r
Model.SARIMA.fit<-arima(Bureau.ts, order = c(1, 0, 2),seasonal = list(order = c(2,1,2), period = 12))
checkresiduals(Model.SARIMA.fit)
```

```r
Box.test(Model.SARIMA.fit$residuals)
```

### ===>le test de Box-Pierce donne un pi_value pour notre nouveau modele sarima qui est supérieur à 0.05 donc c'est un bon modéle parfait

```r
forecast(Model.SARIMA.fit,48)
```
